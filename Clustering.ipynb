```python
from pycaret.clustering import * 
```

```python
clu1 = setup(df, ignore_features=['Type 2'], session_id=123, log_experiment=True, experiment_name='Pokemon')
```

```python
models()
```

```python
kmeans = create_model('kmeans', num_clusters=4)
```

```python
kmodes = create_model('kmodes', num_clusters=4)
```

```python
kmeans_result = assign_model(kmeans)
kmeans_result.head()
```

```python
kmeans_result = assign_model(kmeans)
kmeans_result.head()
```

```python
kmodes_result = assign_model(kmodes)
kmodes_result.head()
```

```python
leg = predict_model(kmeans, data=df1)
leg.head()
```
